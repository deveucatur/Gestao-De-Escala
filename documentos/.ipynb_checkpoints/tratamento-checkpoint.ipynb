```python
import pandas as pd
import json


def semana_by_number(string_sem):
    string_sem = str(string_sem).strip()
    aux = {'Seg': 0,
           'Ter': 1,
           'Qua': 2,
           'Qui': 3,
           'Sex': 4,
           'Sáb': 5,
           'Dom': 6}
    
    if string_sem in aux.keys():
        retorno = aux[string_sem]
    else:
        retorno = None

    return retorno


planil = pd.read_excel('ProjetoHorários.xlsx', sheet_name='Projeto')

cod_linhas_ex = list(set(['{} // {}'.format(str(planil['Projeto - Horários'][x]).strip(), str(planil['Unnamed: 2'][x]).strip()) for x in range(len(planil['Projeto - Horários'])) if str(planil['Projeto - Horários'][x]).strip() not in ('Cód Linha', 'nan')]))

print(len(cod_linhas_ex))
dic_dados = {}
for lin_txt in cod_linhas_ex:
    idx_planil = ['{} // {}'.format(str(planil['Projeto - Horários'][x]).strip(), str(planil['Unnamed: 2'][x]).strip()) for x in range(len(planil['Projeto - Horários']))].index(lin_txt)
    
    aux1 = {
        'LINHA': list(planil['Unnamed: 2'])[idx_planil],
        'HORÁRIO INÍCIO': str(list(planil['Unnamed: 1'])[idx_planil]),
        'FREQUÊNCIA' : [semana_by_number(planil[coln][1]) for coln in ['Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 
                                    'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19']
                                    if str(planil[coln][idx_planil]).strip() == 'x']
       }
        
    #DESCOBRINDO OS SENTIDOS DA LINHAS(SE TEM IDA E VOLTA)
    sentidos_by_lin = list(set([planil['Unnamed: 12'][idx_lin] for idx_lin in range(len(list(planil['Projeto - Horários']))) if str(planil['Projeto - Horários'][idx_lin]).strip() == str(lin_txt.split(' // ')[0]).strip()]))
    
    
    #DADOS SOMENTE DAQUELE SENTIDO
    aux2 = {}
    for sentid in sentidos_by_lin:
        aux3 = {}
        for x in range(len(planil['Projeto - Horários'])):
            if planil['Unnamed: 12'][x] == sentid:
                if str(planil['Projeto - Horários'][x]).strip() == str(lin_txt.split(' // ')[0]).strip():
                    aux3[planil['Unnamed: 3'][x]] = {
                            'SECCIONAMENTO': planil['Unnamed: 4'][x], 
                            'HORÁRIO CHEGADA': str(planil['Unnamed: 5'][x]),
                            'TEMPO DE PARADA': str(planil['Unnamed: 7'][x]),
                            'HORÁRIO SAIDA': str(planil['Unnamed: 9'][x])
                           }

                    aux2[sentid] = aux3
    aux1['PERCURSO'] = aux2
    

    dic_dados[lin_txt.strip()] = aux1
    
    
display(dic_dados)
```

```python
with open('projeto_linhas_horarios.json', 'w') as arquivo_json:
    json.dump(dic_dados, arquivo_json, indent=4)
    
print('ARQUIVO JSON CRIADO')
```
