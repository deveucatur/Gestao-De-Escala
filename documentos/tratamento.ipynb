```python
import pandas as pd
import json

planil = pd.read_excel('ProjetoHorários.xlsx', sheet_name='Projeto')

cod_linhas_ex = list(set([x for x in list(planil['Projeto - Horários']) if str(x).strip() not in ('Cód Linha', 'nan')]))

dic_dados = {}
for lin in cod_linhas_ex:
    idx_planil = [x for x in list(planil['Projeto - Horários'])].index(lin)
    
    aux1 = {'LINHA': list(planil['Unnamed: 2'])[idx_planil],
            'HORÁRIO INÍCIO': str(list(planil['Unnamed: 1'])[idx_planil])}
    
    #DESCOBRINDO OS SENTIDOS DA LINHAS(SE TEM IDA E VOLTA)
    sentidos_by_lin = list(set([planil['Unnamed: 12'][idx_lin] for idx_lin in range(len(list(planil['Projeto - Horários']))) if planil['Projeto - Horários'][idx_lin] == lin]))
    
    #DADOS SOMENTE DAQUELE SENTIDO
    aux2 = {}
    for sentid in sentidos_by_lin:
        aux3 = {}
        for x in range(len(planil['Projeto - Horários'])):
            if planil['Unnamed: 12'][x] == sentid:
                if planil['Projeto - Horários'][x] == lin:
                    aux3[planil['Unnamed: 3'][x]] = {
                            'SECCIONAMENTO': planil['Unnamed: 4'][x], 
                            'HORÁRIO CHEGADA': str(planil['Unnamed: 5'][x]),
                            'TEMPO DE PARADA': str(planil['Unnamed: 7'][x]),
                            'HORÁRIO SAIDA': str(planil['Unnamed: 9'][x])
                           }

                    aux2[sentid] = aux3
    aux1['PERCURSO'] = aux2

    dic_dados[lin] = aux1
    
    
with open('projeto_linhas_horarios.json', 'w') as arquivo_json:
    json.dump(dic_dados, arquivo_json, indent=4)
    
print('ARQUIVO JSON CRIADO')
```
